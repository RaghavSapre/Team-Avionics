```python
import cv2 as cv
from PIL import Image, ImageDraw, ImageFont
from pyzbar.pyzbar import decode
import numpy as np
import matplotlib.pyplot as plt
import pyzbar.pyzbar as pyzbar
```

```python


img = Image.open('./lat longi qr.png')

draw = ImageDraw.Draw(img)
# # font = ImageFont.truetype('Arial.ttf', size=20)  # Set 'arial.ttf' for Windows
# font = ImageFont.truetype(truetype='arial.ttf', size=20)  # Set 'arial.ttf' for Windows
for d in decode(img):
    draw.rectangle(((d.rect.left, d.rect.top), (d.rect.left + d.rect.width, d.rect.top + d.rect.height)),
                   outline=(0, 0, 255), width=3)
    draw.polygon(d.polygon, outline=(0, 255, 0), width=3)
    draw.text((d.rect.left, d.rect.top + d.rect.height), d.data.decode(),
              (255, 0, 0))
    print(d.data.decode())
img.save('./lat longi qr scanned.png')
```

```python
def qrcode(frame):
    try:
        return decode(frame)
    except:
        return []

def draw(frame,qrobj):
    if len(qrobj)==0:
        return frame
    else:
        for obj in qrobj:
            text = obj.data.decode('utf-8')
            points = obj.polygon
            points = np.array(points, np.int32)
            points  = points.reshape((4,1,2))
            cv.polylines(frame,[points],True,(0,255,0),3)
            cv.putText(frame,text,(50,50),cv.FONT_HERSHEY_SIMPLEX,1,(255,0,0),2)
            print(text)
            return frame
        
cap = cv.VideoCapture(0)

while True:
    ret, frame = cap.read()
    qr_obj = qrcode(frame)
    frame = draw(frame,qr_obj)

    cv.imshow('frame', frame)
    if cv.waitKey(1) & 0xFF == ord('q'):
        break
```
